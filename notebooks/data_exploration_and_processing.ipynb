```python
fig,axes = plt.subplots(1,1,figsize=(10, 8))
length_counts = df['pitchLength'].value_counts()
axes.pie(length_counts.values, labels=length_counts.index, autopct='%1.1f%%', startangle=90)
axes.set_title('Distribution of Bowling Lengths')
plt.show()
```

```python
fig1,axes1 = plt.subplots(1,1,figsize=(10, 8))
line_counts = df['pitchLine'].value_counts()
axes1.bar(line_counts.index, line_counts.values, color=['#ff9999', '#66b3ff', '#99ff99'])
axes1.set_title('Distribution of Bowling Lines')
axes1.set_ylabel('Number of Deliveries')
plt.show()

```

```python
fig2,axes2 = plt.subplots(1,1,figsize=(10, 8))
length_line_counts = pd.crosstab(df['pitchLength'], df['pitchLine'])
sns.heatmap(length_line_counts, annot=True, fmt='d', cmap='YlOrBr', ax=axes2)
axes2.set_title('Bowling Length vs Line Combinations')
plt.show()
```

```python
total_fours = df['isFour'].sum()
total_sixes = df['isSix'].sum()
total_wickets = df['isWicket'].sum()
print(f"Total 4s: {total_fours}")
print(f"Total 6s: {total_sixes}") 
print(f"Total wickets: {total_wickets}")
print(f"Total extras (wides + noballs): {df['wides'].sum() + df['noballs'].sum()}")
```

```python
fig3,axes3=plt.subplots(1,1,figsize=(10,8))
wickets_by_length = df.groupby('pitchLength')['isWicket'].mean()*100
axes3.bar(wickets_by_length.index, wickets_by_length.values, color='red', alpha=0.7)
axes3.set_title('Wicket Percentage by Bowling Length')
axes3.set_ylabel('Wicket %')
plt.tight_layout()
plt.show()
```

```python
fig4,axes4=plt.subplots(1,1,figsize=(10,8))
boundary_rate = df.groupby('pitchLength').apply(
    lambda x: ((x['isFour'].sum() + x['isSix'].sum()) / len(x)) * 100
)
axes4.bar(boundary_rate.index, boundary_rate.values, color='blue', alpha=0.7)
axes4.set_title('Boundary Percentage by Bowling Length')
axes4.set_ylabel('Boundary %')
plt.tight_layout()
plt.show()
```

```python
fig5,axes5=plt.subplots(1,1,figsize=(10,8))
avg_runs = df.groupby(['pitchLength', 'pitchLine'])['run'].mean().unstack()
sns.heatmap(avg_runs, annot=True, cmap='YlOrRd', ax=axes5)
axes5.set_title('Average Runs by Bowling Type')
plt.show()
```

```python
fig6,axes6=plt.subplots(1,1,figsize=(10,8))
dot_balls = df.groupby('pitchLength').apply(
    lambda x: (x['run'] == 0).sum() / len(x) * 100
)
axes6.bar(dot_balls.index, dot_balls.values, color='green', alpha=0.7)
axes6.set_title('Dot Ball Percentage by Bowling Length')
axes6.set_ylabel('Dot Ball %')
plt.tight_layout()
plt.show()
```
